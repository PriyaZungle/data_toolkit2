```python
#2 Write a pandas script to find the mean of a specific column in a dataframe.
import pandas as pd

# Sample dataframe
data = {
    'Column1': [10, 20, 30, 40, 50],
    'Column2': [5, 15, 25, 35, 45]
}

df = pd.DataFrame(data)

# Calculate the mean of 'Column1'
mean_column1 = df['Column1'].mean()

print("Mean of Column1:", mean_column1)
```

```python
#3 Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

# Sample data for plotting
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 50]

# Create a scatter plot
plt.scatter(x, y, color='blue', label='Data points')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show a legend
plt.legend()

# Display the plot
plt.show()
```

```python
#  4 How do you calculate the correlation matrix using Seabornand visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataframe
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [7, 8, 9, 10, 11]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
#5 Generate a bar plot using plotly.
import plotly.graph_objects as go

# Sample data for the bar plot
x = ['Category A', 'Category B', 'Category C', 'Category D']
y = [10, 15, 7, 12]

# Create the bar plot
fig = go.Figure(data=[go.Bar(x=x, y=y)])

# Add title and labels
fig.update_layout(
    title='Bar Plot Example',
    xaxis_title='Categories',
    yaxis_title='Values'
)

# Show the plot
fig.show()
```

```python
#6 Create  a data frame and add a new column based on an existing column.
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40]}

df = pd.DataFrame(data)

# Add a new column based on an existing column
df['Age Category'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')

# Display the DataFrame
print(df)
```

```python
#8 create a line plot with multiple lines using matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y1 = np.sin(x)  # First line (sine function)
y2 = np.cos(x)  # Second line (cosine function)
y3 = np.tan(x)  # Third line (tangent function, avoiding large values)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="sin(x)", color='r', linestyle='-')
plt.plot(x, y2, label="cos(x)", color='b', linestyle='--')
plt.plot(x, y3, label="tan(x)", color='g', linestyle=':')

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot using Matplotlib")
plt.legend()

# Limit y-axis to avoid large tangent values
plt.ylim(-2, 2)

# Show the plot
plt.show()
```

```python
#9 generate a pandas dataframe and filter rows where a column value is greater than a threshold.
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 35, 40, 28]}

df = pd.DataFrame(data)

# Define the threshold
threshold = 30

# Filter rows where 'Age' is greater than the threshold
filtered_df = df[df['Age'] > threshold]

# Display the filtered DataFrame
print(filtered_df)
```

```python
#10 create a histogram using seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = np.random.normal(loc=50, scale=15, size=200)  # Normal distribution with mean=50, std=15, size=200

# Create a histogram using seaborn
sns.set_style("whitegrid")  # Set style for better visualization
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=20, kde=True, color='blue')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data")

# Show the plot
plt.show()
```

```python
#11 perform matrix multiplication using numpy.
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = np.dot(A, B)  # OR use A @ B

# Display the matrices and result
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```

```python
#13 create a 3D scatter plot using plotly.
import plotly.graph_objects as go
import numpy as np

# Generate sample data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(size=5, color=z, colorscale='Viridis', opacity=0.8)
)])

# Add title and labels
fig.update_layout(
    title="3D Scatter Plot",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Show the plot
fig.show()
```
